```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset into DataFrame
df = pd.read_csv(r"/Users/saifpatel/Downloads/NullClass Internship/job_descriptions.csv")
```

```python
# Display the first few rows of the dataframe
df.head()
# Get a summary of the dataframe
df.info()
```

```python
# Identifying missing values
df.isnull().sum()
```

```python
# Fill missing values with an empty string
df['Company Profile'].fillna('', inplace=True)
```

```python
# Standardize date formats
df['Job Posting Date'] = pd.to_datetime(df['Job Posting Date'], format='%Y-%m-%d')
```

```python
# Rename column names
df.rename(columns={'Job Id' : 'Job_Id', 'Salary Range' : 'Salary_Range', 'Work Type' : 'Work_Type', 'Company Size' : 'Company_Size', 
                   'Job Posting Date' : 'Job_Posting_Date', 'Contact Person' : 'Contact_Person', 'Job Title' : 'Job_Title', 
                   'Job Portal' : 'Job_Portal', 'Job Description' : 'Job_Description', 'Company Profile' : 'Company_Profile'}, inplace=True)
```

```python
# Function to convert salary range to numeric values
def parse_salary_range(salary_range):
    min_salary, max_salary = salary_range.split('-')
    min_salary = int(min_salary.replace('$', '').replace('K', '')) * 1000
    max_salary = int(max_salary.replace('$', '').replace('K', '')) * 1000
    return min_salary, max_salary

# Apply the function to create new columns
df[['Salary_Min', 'Salary_Max']] = df['Salary_Range'].apply(lambda x: pd.Series(parse_salary_range(x)))

# Optional: Calculate the average salary
df['Salary_Avg'] = (df['Salary_Min'] + df['Salary_Max']) / 2
```

```python
import re

# Function to convert experience range to numerical (average experience years)
def convert_experience(experience):
    match = re.match(r"(\d+) to (\d+) Years", experience)
    if match:
        return (int(match.group(1)) + int(match.group(2))) / 2
    return None

# Function to convert salary range to numerical (average salary in thousands)
def convert_salary(salary):
    match = re.match(r"\$(\d+)K-\$(\d+)K", salary)
    if match:
        return (int(match.group(1)) + int(match.group(2))) / 2
    return None

# Apply the conversion functions
df['Experience'] = df['Experience'].apply(convert_experience)
df['Salary_Range'] = df['Salary_Range'].apply(convert_salary)

df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Generate summary statistics for numerical columns
summary_statistics = df.describe()

# Get value counts for categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
value_counts = {col: df[col].value_counts() for col in categorical_columns}

missing_values, summary_statistics, value_counts
```
